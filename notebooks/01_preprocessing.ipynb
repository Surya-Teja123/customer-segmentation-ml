## Preprocessing

```python
# Z_CostContact and Z_Revenue are constant columns, ID is just a row identifier
df = df.drop(columns=['ID', 'Z_CostContact', 'Z_Revenue'])
print("Shape after dropping:", df.shape)
```

```python
# Fill 24 missing Income values with median (better than mean, ignores outliers)
df['Income'] = df['Income'].fillna(df['Income'].median())
print("Missing values remaining:", df.isnull().sum().sum())
```

#### Creating New Features

```python
import datetime

# Age of customer
current_year = datetime.datetime.now().year
df['Age'] = current_year - df['Year_Birth']

# Total amount spent across all product categories
df['TotalSpend'] = (df['MntWines'] + df['MntFruits'] + df['MntMeatProducts'] +
                    df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds'])

# Total number of purchases across all channels
df['TotalPurchases'] = (df['NumDealsPurchases'] + df['NumWebPurchases'] +
                        df['NumCatalogPurchases'] + df['NumStorePurchases'])

# Total number of children at home
df['TotalChildren'] = df['Kidhome'] + df['Teenhome']

# How many campaigns accepted in total
df['TotalCampaignsAccepted'] = (df['AcceptedCmp1'] + df['AcceptedCmp2'] +
                                 df['AcceptedCmp3'] + df['AcceptedCmp4'] +
                                 df['AcceptedCmp5'])

# Customer enrollment year — fixed format
df['EnrollmentYear'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True).dt.year

print("New features added. Shape:", df.shape)
df[['Age', 'TotalSpend', 'TotalPurchases', 'TotalChildren', 
    'TotalCampaignsAccepted', 'EnrollmentYear']].head()
```

### Droping original columns which are no longer needed

```python
df = df.drop(columns=['Year_Birth', 'Dt_Customer'])
print("Shape:", df.shape)
```

```python
# Check unique values first
print("Education:", df['Education'].unique())
print("Marital_Status:", df['Marital_Status'].unique())
```

### Clean Marital Status

```python
# Group messy values into clean categories
df['Marital_Status'] = df['Marital_Status'].replace({
    'Alone': 'Single',
    'Absurd': 'Single',
    'YOLO': 'Single'
})

print("Marital_Status cleaned:", df['Marital_Status'].unique())
```

### Encode Education

```python
# Map education to ordinal numbers (ordered low to high)
education_order = {
    'Basic': 0,
    '2n Cycle': 1,
    'Graduation': 2,
    'Master': 3,
    'PhD': 4
}

df['Education'] = df['Education'].map(education_order)
print("Education encoded:\n", df['Education'].value_counts())
```

### Encode Marital Status

```python
# One-hot encode marital status (no natural order so we don't use numbers)
df = pd.get_dummies(df, columns=['Marital_Status'], drop_first=True)
print("Shape after encoding:", df.shape)
print("New columns:", [col for col in df.columns if 'Marital' in col])
```

### Remove Outliers

```python
# Age above 90 and Income above 600000 are likely data errors
print("Age max before:", df['Age'].max())
print("Income max before:", df['Income'].max())

df = df[df['Age'] < 90]
df = df[df['Income'] < 600000]

print("\nAge max after:", df['Age'].max())
print("Income max after:", df['Income'].max())
print("Shape after removing outliers:", df.shape)
```

### Scale Numerical Features

```python
from sklearn.preprocessing import StandardScaler

# Columns to scale
scale_cols = ['Income', 'Age', 'TotalSpend', 'TotalPurchases', 
              'Recency', 'NumWebVisitsMonth', 'TotalChildren',
              'TotalCampaignsAccepted', 'EnrollmentYear']

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[scale_cols] = scaler.fit_transform(df[scale_cols])

print("Scaling done!")
df_scaled[scale_cols].describe().round(2)
```

### Save Cleaned data

```python
import os

# Save both versions — we need unscaled for EDA, scaled for ML
os.makedirs('../data', exist_ok=True)
df.to_csv('../data/cleaned_data.csv', index=False)
df_scaled.to_csv('../data/scaled_data.csv', index=False)

print("✅ Preprocessing complete!")
print("cleaned_data.csv → use for EDA and dashboards")
print("scaled_data.csv  → use for ML models and clustering")
```
